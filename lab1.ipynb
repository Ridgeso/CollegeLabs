```python
import cv2
import matplotlib.pylab as pyplot

import math
```

```python
imageOrg = cv2.imread("lab1.jpg")
imageOrg = cv2.cvtColor(imageOrg, cv2.COLOR_BGR2RGB)
```

```python
image = imageOrg.copy()
```

```python
new_image = image[90:140, 105:155]
pyplot.imshow(new_image)
```

```python
resized_im = cv2.resize(new_image, (new_image.shape[0] * 3, new_image.shape[1] * 3))
pyplot.imshow(resized_im)
```

```python
new_size = (int(resized_im.shape[0] / 0.86602540378 + resized_im.shape[1] / 0.5), int(resized_im.shape[0] / 0.86602540378 + resized_im.shape[1] / 0.5))
center = ((resized_im.shape[0] + new_size[0]) / 2, (resized_im.shape[1] - new_size[1]) / 2)
rotate_matrix = cv2.getRotationMatrix2D(center=center, angle=30, scale=1)
rotated_image = cv2.warpAffine(src=resized_im, M=rotate_matrix, dsize=new_size)
pyplot.imshow(rotated_image)
```

```python
named_im = cv2.rectangle(resized_im.copy(), (12, 12), (140, 140), (244, 21, 21), 1)
named_im = cv2.putText(named_im, 'Graka', (30, 140), cv2.FONT_HERSHEY_SIMPLEX, 1, (163, 163, 174), 1, cv2.LINE_AA)
pyplot.imshow(named_im)
```
